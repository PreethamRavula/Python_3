#### Step 1: Decode Vishal's Message
In the cell below, use your Python skills to decode Vishal's message and print the result. Hint: you can account for shifts that go past the end of the alphabet using the modulus operator, but I'll let you figure out how!

```python
#Creating the Alphabets String
alphabets = "abcdefghijklmnopqrstuvwxyz"

#The coded Caesar Cipher message
message = "xuo jxuhu! jxyi yi qd unqcfbu ev q squiqh syfxuh. muhu oek qrbu je tusetu yj? y xefu ie! iudt cu q cuiiqwu rqsa myjx jxu iqcu evviuj!"

# Splitting the given message with a space
message_split = message.split(" ")
print(message_split)

#Assigning the positions to split message using alphabets string
positions = []
for word in message_split:
    position_word = []
    for alphabet in word:
        position = alphabets.find(alphabet)
        position_word.append(position)
    positions.append(position_word)
print(positions)

#decoding the letters with offset and storing them in a list
shifted_words_list = []
count = 0
for word_pos_list in positions:
    shifted_char_list = []
    for char_pos in word_pos_list:
        if char_pos in range(len(alphabets)):
            shifted_char_list.append(alphabets[(char_pos + 10)%26])
            count += 1
        else:
            shifted_char_list.append(message[count])
            count += 1
    count += 1
    shifted_words_list.append(shifted_char_list)
print(shifted_words_list)

# Converting the original message list to original sentence string
original_words = ""
original_words_list = []
for original_word in shifted_words_list:
    original_words = "".join(original_word)
    original_words_list.append(original_words)
print(original_words_list)
original_sentence = " ".join(original_words_list)
print("\nDeocoded message:\n\n", original_sentence)
    
        
```

#### Step 5: The Vigenère Cipher

Great work! While you were working on the brute force cracking of the cipher, Vishal sent over another letter. That guy is a letter machine!

            Salutations! As you can see, technology has made brute forcing simple ciphers like the Caesar Cipher extremely easy, and us crypto-enthusiasts have had to get more creative and use more complicated ciphers. This next cipher I'm going to teach you is the Vigenère Cipher, invented by an Italian cryptologist named Giovan Battista Bellaso (cool name eh?) in the 16th century, but named after another cryptologist from the 16th century, Blaise de Vigenère.
            
           The Vigenère Cipher is a polyalphabetic substitution cipher, as opposed to the Caesar Cipher which was a monoalphabetic substitution cipher. What this means is that opposed to having a single shift that is applied to every letter, the Vigenère Cipher has a different shift for each individual letter. The value of the shift for each letter is determined by a given keyword.
           
           Consider the message
           
               barry is the spy

           If we want to code this message, first we choose a keyword. For this example, we'll use the keyword
           
               dog
               
           Now we use the repeat the keyword over and over to generate a _keyword phrase_ that is the same length as the message we want to code. So if we want to code the message "barry is the spy" our _keyword phrase_ is "dogdo gd ogd ogd". Now we are ready to start coding our message. We shift the each letter of our message by the place value of the corresponding letter in the keyword phrase, assuming that "a" has a place value of 0, "b" has a place value of 1, and so forth. Remember, we zero-index because this is Python we're talking about!

                        message:       b  a  r  r  y    i  s   t  h  e   s  p  y
                
                 keyword phrase:       d  o  g  d  o    g  d   o  g  d   o  g  d
                 
          resulting place value:       4  14 15 12 16   24 11  21 25 22  22 17 5
      
            So we shift "b", which has an index of 1, by the index of "d", which is 3. This gives us an place value of 4, which is "e". Then continue the trend: we shift "a" by the place value of "o", 14, and get "o" again, we shift "r" by the place value of "g", 15, and get "x", shift the next "r" by 12 places and "u", and so forth. Once we complete all the shifts we end up with our coded message:
            
                eoxum ov hnh gvb
                
            As you can imagine, this is a lot harder to crack without knowing the keyword! So now comes the hard part. I'll give you a message and the keyword, and you'll see if you can figure out how to crack it! Ready? Okay here's my message:
            
                dfc aruw fsti gr vjtwhr wznj? vmph otis! cbx swv jipreneo uhllj kpi rahjib eg fjdkwkedhmp!
                
            and the keyword to decode my message is 
            
                friends
                
            Because that's what we are! Good luck friend!
           
And there it is. Vishal has given you quite the assignment this time! Try to decode his message. It may be helpful to create a function that takes two parameters, the coded message and the keyword and then work towards a solution from there.

**NOTE:** Watch out for spaces and punctuation! When there's a space or punctuation mark in the original message, there should be a space/punctuation mark in the corresponding repeated-keyword string as well! 

```python
# Vigenere Cipher decoder function:
def vigenere_decoder(code_message, keyword):
    # Setting the Alphabets refernce list:
    alphabets = "abcdefghijklmnopqrstuvwxyz"
    split_code_message = code_message.split()
    print(split_code_message)
    
    # Splitting the coded message into individual words list:
    split_code_words = []
    for code_word in split_code_message:
        split_code_chars = []
        for code_word_char in code_word:
            split_code_chars.append(alphabets.find(code_word_char)) 
        split_code_words.append(split_code_chars)
    print("\n", split_code_words)
    
    # Generating the position value list of the coded message:
    count = 0
    decoded_phrase_pos_list = []
    for word_pos in split_code_words:
        decoded_word_pos_list = []
        for char_pos in word_pos:
            if char_pos != -1:
                decoded_word_pos_list.append(char_pos - alphabets.find(keyword[count % len(keyword)]))
                count += 1
            else:
                decoded_word_pos_list.append(-1)
        decoded_phrase_pos_list.append(decoded_word_pos_list)
    print("\n", decoded_phrase_pos_list)
    
    # Generating the updated position value list with the reference keyword:
    decoded_phrase_list = []
    num = 0
    for decoded_word in decoded_phrase_pos_list:
        decoded_word_list = []
        for decoded_char in decoded_word:
            if decoded_char != -1:
                decoded_word_list.append(alphabets[decoded_char%26])
                num += 1
            else:
                decoded_word_list.append(code_message[num])
                num += 1
        num += 1
        decoded_phrase_list.append(decoded_word_list)
    print("\n", decoded_phrase_list)
    
    # Converting the decoded list phrase to a string using join method:
    decoded_words_list = []
    for real_word in decoded_phrase_list:
        decoded_words_list.append("".join(real_word))
    print("\n", decoded_words_list)
    
    decoded_phrase = " ".join(decoded_words_list)
    print("\n The final decoded phrase is:\n\n {}".format(decoded_phrase))
    
    return decoded_phrase
         
vigenere_decoder("dfc aruw fsti gr vjtwhr wznj? vmph otis! cbx swv jipreneo uhllj kpi rahjib eg fjdkwkedhmp!", "friends")
```

```python
# Vigenere Cipher Coder function:
def vigenere_coder(actual_message, keyword):
    # Setting the Alphabets refernce list:
    alphabets = "abcdefghijklmnopqrstuvwxyz"
    split_actual_message = actual_message.split()
    print(split_actual_message)
    
    # Splitting the coded message into individual words list:
    split_actual_words = []
    for actual_word in split_actual_message:
        split_actual_chars = []
        for actual_word_char in actual_word:
            split_actual_chars.append(alphabets.find(actual_word_char)) 
        split_actual_words.append(split_actual_chars)
    print("\n", split_actual_words)
    
    # Generating the position value list of the coded message:
    count = 0
    actual_phrase_pos_list = []
    for actual_word_pos in split_actual_words:
        actual_word_pos_list = []
        for actual_char_pos in actual_word_pos:
            if actual_char_pos != -1:
                actual_word_pos_list.append(actual_char_pos + alphabets.find(keyword[count % len(keyword)]))
                count += 1
            else:
                actual_word_pos_list.append(-1)
        actual_phrase_pos_list.append(actual_word_pos_list)
    print("\n", actual_phrase_pos_list)
    
    # Generating the updated position value list with the reference keyword:
    coded_phrase_list = []
    num = 0
    for coded_word in actual_phrase_pos_list:
        coded_word_list = []
        for coded_char in coded_word:
            if coded_char != -1:
                coded_word_list.append(alphabets[coded_char%26])
                num += 1
            else:
                coded_word_list.append(actual_message[num])
                num += 1
        num += 1
        coded_phrase_list.append(coded_word_list)
    print("\n", coded_phrase_list)
    
    # Converting the decoded list phrase to a string using join method:
    coded_words_list = []
    for real_word in coded_phrase_list:
        coded_words_list.append("".join(real_word))
    print("\n", coded_words_list)
    
    coded_phrase = " ".join(coded_words_list)
    print("\n The final decoded phrase is:\n\n {}".format(coded_phrase))
    
    return coded_phrase

vigenere_coder("You keep throwing your little challenges at me and i keep knocking them right out of the park, keep it coming!", "loser")
```
